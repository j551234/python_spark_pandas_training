```python
## python sublist

testList = []  
for i in range(100):
    testList.append(i)
testB = testList[46:88]
for number in testB:
    print(number)

```

```python
## list reverse

numbers = testList.copy()
for number in numbers:
    print(number)
numbers.reverse()
print("now is reverse")
for number in numbers:
    print(number)
```
