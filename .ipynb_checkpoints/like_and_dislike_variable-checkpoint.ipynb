```python
import pandas as pd
```

```python
def get_category(file_name):
    df = pd.read_json(file_name)
    data = pd.json_normalize(df['items'])
    return data[['id','snippet.title']]
```

```python

df = pd.read_csv("/home/james/python_practice/archive/CAvideos.csv")
tag = get_category("/home/james/python_practice/archive/CA_category_id.json")

## 轉型 方便 Join 時候結合
df['category_id']=df['category_id'].astype(int)
tag['id']=tag['id'].astype(int)

## join 語法 指定join欄位
result = pd.merge(df,tag,left_on ='category_id',right_on='id')
display(result)
```

```python
data = result[['snippet.title','dislikes','likes']]

data = data.rename({'snippet.title': 'category'}, axis=1) 
data=data.assign(**{"ratio":data["likes"]/(data["likes"]+data["dislikes"])})
## 聚合 category 的dislike數量
data['ratio_var']=data['ratio'].var()
## 排序 dislike 數量

# group = group.sort_values(by='dislikes',ascending=False)
display(new_data)
```
